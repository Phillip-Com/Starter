```
import random
import numpy as np
print(np.__version__)
```

```
# Variables
played_board = ["   |","   |","   \n---|---|---\n",
                 "   |","   |","   \n---|---|---\n",
                 "   |","   |","   \n"]
selection_board = [" 1 |"," 2 |"," 3 \n---|---|---\n",
                     " 4 |"," 5 |"," 6 \n---|---|---\n",
                     " 7 |"," 8 |"," 9 \n"]
board = np.zeros((3,3), dtype=int)
num_convert_row = [0,1,2,
                   0,1,2,
                   0,1,2]
num_convert_col = [0,0,0,
                   1,1,1,
                   2,2,2]
blank = "   "
player = blank
comp = blank
options = [1,2,3,4,5,6,7,8,9]
turn = 0
counter = 1
point = 0
```

```
# Function to find available position
def availability(play):
    row, col = np.where(play == 0)
    return list(zip(row, col))

# Function to check if game is won
def won_game(play, piece):
    # Loop through the board to see if game board has winning position
    for i in range(3):
        if all(play[i, :] == piece) or all(play[:, i] == piece):
            return True
    if all(play.diagonal() == piece) or all(play.diagonal() == piece):
            return True
    else:
        return False

# Function to find a random move
def random_move(play):
    # Find available space
    moves = availability(play)

    # Return random position
    if moves:
        return random.choice(moves)

    # Return none if no positions are available
    return None

# Function to find the best move
def smart_move(play, comp_piece, player_piece):
    # Check if computer can win in next move
    for r, c in availability(play):
        play_copy = play.copy()
        play_copy[r,c] = comp_piece
        if won_game(play_copy, comp_piece):
            print("Hit Comp Win")
            return r,c

    # Check if player can win in next move
    for r, c in availability(play):
        play_copy = play.copy()
        play_copy[r,c] = player_piece
        if won_game(play_copy, player_piece):
            print("Hit Player Win")
            return r,c

    # Try to take center
    if play[1,1] == 0:
        print("Hit Center Take")
        return 1,1

    # Try to take corner
    corners = [(0,0),(0,2),(2,0),(2,2)]
    available_corners = [pos for pos in corners if pos in availability(play)]
    if available_corners:
        print("Hit Corner Take")
        return random.choice(available_corners)

    # Pick random move if nothing else
    print("Hit Random")
    return random_move(play)
```

```
# Output directions for selection Xs or Os
print("Input O if you want to play O, Input X if you want to play X")
user = ""
while True:
    # Gets user input for their choice
    user = input("Enter your choice: ").lower()

    # Catch invalid inputs
    if user == "o" or user == "x":
        break
    else:
        print("Invalid Input")

# Display message for what they chose
print(f"You have chosen to play as ", {user})

# Set up game variables
if user == "o":
    turn = 0
    player = " O "
    comp = " X "
else:
    turn = 1
    player = " X "
    comp = " O "

# Begin game loop with 9 rounds
while counter < 10:

    # Computer Turn
    if turn == 0:
        # Generate position for Comp
        print(smart_move(board, -1, 1))
        x, y = smart_move(board, -1, 1)
        if y is None:
            counter = 9
            break
        if y == 0:
            point = x + y
        if y == 1:
            point = x + y + 2
        if y == 2:
            point = x + y + 4
        randomNum = point + 1

        # Update boards for display and logic
        board[x][y] = -1
        played_board[point] = played_board[point].replace("   ", comp)
        selection_board[point] = selection_board[point].replace(f" {randomNum} ", blank)

        # Remove selected position from future choices
        options.remove(randomNum)

        # Print what was selected and display new board
        print(f"Putting{comp}in slot {randomNum}")
        display1 = "".join(played_board)
        print(display1)

        # Update turn to give the player a turn
        turn = 1

    #Player Turn
    else:
        # Allow user to input a choice for play and display the choices
        print("Choose a location to put an", player)
        display2 = "".join(selection_board)
        print(display2)

        # Take input from user and create an index selection variable
        selection = int(input("Input number for your selection: "))
        point = selection - 1

        # Update boards for display and logic
        board[num_convert_row[point]][num_convert_col[point]] = 1
        played_board[point] = played_board[point].replace("   ", player)
        selection_board[point] = selection_board[point].replace(f" {selection} ", blank)

        # Remove selected position from future choices
        options.remove(selection)

        # Print what was selected
        print(f"Putting{player}in slot {selection}")

        # Update turn to give the computer a turn
        turn = 0

    # If minimum amount of turns for win, check for a winning position
    if counter >= 5:

        # If last turn was computer
        if turn == 1:

            # check if computer has winning position
            if won_game(board, -1):
                # Display computer has won and the winning board
                print(f"{comp}Wins!")
                display2 = "".join(played_board)
                print(display2)

                # Update counter to exit game loop
                counter = 10
        # If last turn was a player
        if turn == 0:

            # Check if player has winning position
            if won_game(board, 1):
                # Display player has won and the winning board
                print(f"{player}Wins!")
                display2 = "".join(played_board)
                print(display2)

                # Update counter to exit game loop
                counter = 10

    # If on last round then it must be a draw
    if counter == 9:
        # Display drawn game
        print("Draw!")
        display2 = "".join(played_board)
        print(display2)

    # Update counter to move onto next turn
    counter += 1
```
