```
import random
```

```
list1 = ["   ","|","   ","|","   ","\n",
         "---","|","---","|","---","\n",
         "   ","|","   ","|","   ","\n",
         "---","|","---","|","---","\n",
         "   ","|","   ","|","   ","\n"]
x = " X "
o = " O "
options = [0,2,4,12,14,16,24,26,28]
turn = 0
counter = 0
```

```
display = "".join(list1)
print(display)

while counter < 9:
    randomNum = random.choice(options)
    options.remove(randomNum)

    print("Element Selected: ")
    print(randomNum)

    if turn == 0:
        list1[randomNum] = o
        turn = 1
    else:
        list1[randomNum] = x
        turn = 0

    display = "".join(list1)
    print(display)

    

    counter += 1
```
